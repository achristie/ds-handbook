```python
def f(x, y):
  return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
```

```python
Z = f(X,Y)
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
```

```python
contours = plt.contour(X,Y,Z, 3, colors='black')
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy', interpolation='gaussian', aspect='equal')

plt.colorbar()
```

```python
import datetime
```

```python
def convert_time(s):
  h, m, s = map(int, s.split(":"))
  return datetime.timedelta(hours=h, minutes=m, seconds=s)
```

```python
df = pd.read_csv("data/marathon-data.csv", converters={"split": convert_time, 'final': convert_time})
```

```python
df.head()
```

```python
df.info()
```

```python
df['split_sec'] = df['split'].view(int) / 1e9
df['final_sec'] = df['final'].view(int) / 1e9
df.head()
```

```python
g = sns.jointplot(x='split_sec', y='final_sec', data=df, kind='hex')
g.ax_joint.plot(np.linspace(4000, 16000), np.linspace(8000, 32000), ':k')
```

```python
df['split_frac'] = 1 - 2 * df['split_sec'] / df['final_sec']
```

```python
df.head()
```

```python
sns.displot(df['split_frac'], kde=False)
plt.axvline(0, color='k', linestyle='--')
```

```python
g = sns.PairGrid(df, vars=['age', 'split_sec', 'final_sec', 'split_frac'], hue='gender', palette='RdBu_r')
g.map(plt.scatter, alpha=.7)
g.add_legend()
```

```python
sns.kdeplot(df.split_frac[df.gender=="M"], label='men', fill=True)
sns.kdeplot(df.split_frac[df.gender=="W"], label='women', fill=True)
```

```python
sns.violinplot(x='gender', y='split_frac', data=df)
```

```python
df['age_dec'] = df.age.map(lambda age: 10 * (age // 10))
df.head()
```

```python
sns.violinplot(x="age_dec", y="split_frac", data=df, hue='gender')
```

```python
g = sns.lmplot(x='final_sec', y='split_frac', col='gender', data=df, markers='.')
g.map(plt.axhline, y=0, color='k', ls=':')
```
